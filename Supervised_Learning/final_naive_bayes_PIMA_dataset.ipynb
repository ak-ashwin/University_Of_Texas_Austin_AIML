### Experiment 3 - Base Model with Data Processing without stratify


```python
df.hist(figsize=(20,30))
```

```python
df2 = df.copy(deep=True)
df2.loc[df2['Plas']==0,'Plas'] = np.nan
df2.loc[df2['Pres'] ==0,'Pres'] = np.nan
df2.loc[df2['mass']==0,'mass'] = np.nan
df2.loc[df2['skin']==0,'skin'] = np.nan
```

#### Replacing all zero values with NAN except the column "Test"

```python
df2['Plas'].fillna(value=df['Plas'].mean(),inplace=True)
df2['Pres'].fillna(value=df['Pres'].mean(),inplace=True)
df2['mass'].fillna(value=df['mass'].mean(),inplace=True)
df2['skin'].fillna(value=df['skin'].median(),inplace=True)
```

#### Replaced missing values with mean and median values for their respective columns

```python
sns.pairplot(df2,hue='class')
```

```python
X = df2.drop(['class'],axis=1)
y = df2['class']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)

classifier = GaussianNB()
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print (cm)
print(f1_score(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

#### Observation 3:

    Accuracy score : 76.62%
    F1 Score : 67.07%

### Experiment 5 - Standard scaler instead of minmaxscaler as there are no negative values 

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1,stratify=y)
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python


classifier = GaussianNB()
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print (cm)
print(f1_score(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

### Observation 5:

Accuracy score : 70.56% F1 Score : 56.41%
